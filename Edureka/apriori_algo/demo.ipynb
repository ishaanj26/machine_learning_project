```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
df=pd.read_excel('Online_Retail.xlsx')
```

```python
df.head()
df.info()
```

```python
df['Description'] = df['Description'].str.strip()
#Removes leading and trailing spaces from every value in the 'Description' column.

df.dropna(subset=['InvoiceNo'], inplace=True)
#Removes rows where the 'InvoiceNo' column has missing (NaN) values.

df['InvoiceNo'] = df['InvoiceNo'].astype(str)
#Converts all values in the 'InvoiceNo' column to strings.

df = df[~df['InvoiceNo'].str.contains('C')]
#Filters out rows where 'InvoiceNo' contains the letter 'C' (often used to indicate canceled transactions).
```

```python
df.info()
```

```python
basket=(df[df['Country'] == 'France']
         .groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[(rules['lift'] >= 6) & (rules['confidence'] >= 0.8)]
```
