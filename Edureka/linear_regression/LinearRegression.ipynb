```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
cars = pd.read_csv("Car_sales.csv")
#print(cars.head())
print(cars.columns)
```

```python
# Ensure numeric and drop missing values
cars['Horsepower'] = pd.to_numeric(cars['Horsepower'], errors='coerce')
cars['Price in thousands'] = pd.to_numeric(cars['Price in thousands'], errors='coerce')
data = cars[['Horsepower', 'Price in thousands']].dropna()
x = data['Horsepower'].values.reshape(-1, 1)
y = data['Price in thousands'].values.reshape(-1, 1)

reg = LinearRegression()
reg.fit(x, y)

print(reg.coef_[0][0])

prediction = reg.predict(x)

plt.figure(figsize=(16, 8))
plt.scatter(
    data['Horsepower'],
    data['Price in thousands'],
    c='black'
)

plt.plot(
    data['Horsepower'],
    prediction,
    c='red',
    linewidth=2)
plt.xlabel('Horsepower')
plt.ylabel('Price in thousands')
plt.show()

```
