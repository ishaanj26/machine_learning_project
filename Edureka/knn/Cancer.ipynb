```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
```

```python
cancer.keys()
```

```python
df_feat= pd.DataFrame(cancer.data, columns=cancer.feature_names)
df_feat.info()
```

```python
df_target= pd.DataFrame(cancer.target, columns=['Cancer'])
df_target.head()
```

```python
df_feat.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
scaler.fit(df_feat)
scaled_features= scaler.transform(df_feat)
```

```python
df_feat_Scaled= pd.DataFrame(scaled_features, columns=df_feat.columns)
df_feat_Scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_features, np.ravel(df_target), test_size=0.3, random_state=105)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
#Choosing a k value
error_rate = []
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate vs K Value')
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
#with k=21
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```
